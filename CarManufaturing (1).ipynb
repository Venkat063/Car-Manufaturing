#DATA VISUALIZATION

```python
plt.title('Car Model By Prices')
plt.xlabel('CAR PRICES')
plt.ylabel('CAR MODEL')
plt.plot(df['Price'],df['CarModel'])
plt.grid()
plt.show()
```

#BIVARIATE

```python
plt.title('CarModeels By Speed')
plt.plot(df['CarModel'],df['TopSpeed'],color='yellow')
plt.xlabel('Car Model')
plt.ylabel('TopSpeed')
plt.xticks(rotation="vertical")
plt.grid()
plt.show()
```

#Multivirate & Legend

```python
plt.title('Types of Engines & Horse Power')
plt.plot(df['EngineType'],df['Horsepower'],label='Horsepower')
plt.plot(df['EngineType'],df['Torque'],label='Torque')
plt.xlabel('EngineType')
plt.ylabel('Horsepower & Torque')
plt.legend()
plt.grid()
plt.show()
```

#Dotted lines & line width

```python
plt.title('CarModeels By Speed')
plt.plot(df['CarModel'],df['TopSpeed'], linestyle='dotted')
plt.xlabel('Car Model')
plt.ylabel('TopSpeed')
plt.xticks(rotation="vertical")
plt.grid()
plt.show()
```

```python
plt.title('Car Model By Prices')
plt.xlabel('CAR PRICES')
plt.ylabel('CAR MODEL')
plt.plot(df['Price'],df['CarModel'],linewidth=2, marker='*', color='green')
plt.grid()
plt.show()
```

#scatter

```python
plt.title('Car Model By Prices')
plt.xlabel('CAR PRICES')
plt.ylabel('CAR MODEL')
plt.scatter(df['Price'],df['CarModel'],linewidth=2, marker='*', color='green')
plt.grid()
plt.show()
```

#Bar Graph

```python
plt.title('Car Model By Seating')
a=df['CarModel']
b=df['SeatingCapacity']
plt.xlabel('CAR MODEL')
plt.ylabel('CAR Seating Capacity')
plt.xticks(rotation="vertical")
plt.bar(a,b,color='orange')
plt.show()
```

#HORIZONTAL BAR GRAPH

```python
plt.title('Car Model By Seating')
a=df['CarModel']
b=df['SeatingCapacity']
plt.xlabel('CAR MODEL')
plt.ylabel('CAR Seating Capacity')
plt.barh(a,b,color='orange')
plt.show()
```

#PieChart
